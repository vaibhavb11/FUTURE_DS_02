```python
import pandas as pd

df = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\DA Project\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head() # Preview first few records
df.tail()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df)
```

```python
df = df.dropna()
df.info()
```

```python
binary_cols = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

print(df[binary_cols].head())
```

```python
df['TenureGroup'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 24, 36, 48, 60, 72],
    labels=['0-1 yr','1-2 yr','2-3 yr','3-4 yr','4-5 yr','5-6 yr']
)
print(df[['tenure', 'TenureGroup']].head(10))
```

```python
df['MonthlyChargeGroup'] = pd.cut(
    df['MonthlyCharges'],
    bins=[0, 35, 70, 100, 150],
    labels=['Low', 'Medium', 'High', 'Very High']
)
print(df[['MonthlyCharges', 'MonthlyChargeGroup']].head(10))
```

```python
with pd.ExcelWriter(r"C:\Users\HP\OneDrive\Desktop\DA Project\Cleaned_telcon.xlsx", engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='CleanedData', index=False)
print("Successfully upload Data")
```
